In this notebook we will use my functions from index_functions.py to measure spectroscopic indices from a spectrum of a galaxy. The indices in question H$\alpha$, H$\beta$ and D4000 allow us to gain some understanding of the age of the stars that make up the galaxy. H$\alpha$ and H$\beta$ are equivalent width measurements, which is a measure of the strength of an absorption line (in this case the lines are H$\alpha$, H$\beta$) relative to the background continuum - the light being emitted by the galaxy. D4000 is a ratio of two spectral regions that sit either side of a dense cluster of Hydrogen absorption lines called the Balmer break.

For all of these measurements, a stronger index means a younger stellar population as there must be lots of young, Hydrogen-rich stars.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import index_functions as f
```

First, we will load in the spectrum of the galaxy in question and plot it to see what we are dealing with. As with other spectra we have dealt with in these examples, column 0 is wavelength in angstroms and column 1 is flux.

```python
spec = np.genfromtxt('Data/galaxy_spec.txt',comments='#')

plt.plot(spec[:,0], spec[:,1])
plt.xlabel("Wavelength (A)")
plt.ylabel("Flux (erg/s/cm2)")
plt.show()
```

In the plot above we can see the strong H$\alpha$ absorption line at 6563A, a slightly weaker line at 4582A which is H$\beta$, and a dense clustering of absorption lines around 4000A which is the D4000 region.
