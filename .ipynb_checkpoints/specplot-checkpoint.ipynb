# specplot

A simple function to plot a single spectrum within a desired wavelength range using matplotlib. The spectrum needs to be a two-column text/csv/ascii file with wavelength in the x-axis and flux in the y-axis

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def plotspec(spec, z=0, wav_low = 0, wav_high = 0, title = "", xlabel = "", ylabel = ""):
    
    #Correct the wavelength axis for redshift, z. If no value is given,
    #this will divide by 1 and have no effect
    spec[:,0] /= 1 + z
    
    #Trim the wavelength range if limits given
    if wav_low != 0:
        spec = spec[spec[:,0]>wav_low]
    if wav_high != 0:
        spec = spec[spec[:,0]<wav_high]
    
    #Plot the spectrum
    plt.plot(spec[:,0], spec[:,1], lw=1)
    
    #Add title and axis labels if given
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.show()
```

As an example I am plotting the spectrum of a tidal disruption event which has wavelength in units of Angstroms and flux in units of erg s$^{-1}$cm$^{-2}$A$^{-1}$. I am setting an upper wavelength limit of 7000 Angstroms. 

```python
s1 = np.genfromtxt('Data/TDE_spec.txt')
z =  0.04573 #redshift of host galaxy

plotspec(s1, z, wav_high = 7000, xlabel = r'Rest Wavelength ($\AA$)', ylabel = r'Flux (erg s$^{-1}$cm$^{-2}\AA^{-1}$)')
```
