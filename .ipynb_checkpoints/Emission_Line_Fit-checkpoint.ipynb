# Emission Line Fitting

These functions use lmfit to fit models to emission lines. The goal is to be able to measure physical properties such as the width of the emission line and its total flux. 

The first function will fit a single Gaussian function to the emission line a polynomial function to the continuum emission (the underlying shape of the spectrum). The second will fit several Gaussians which is often necessary when emission lines overlap.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

from lmfit.models import GaussianModel
from lmfit.models import PolynomialModel
```

```python
def fit_gauss(s, date, line, wave, cont, plot):

    """
    
    This function fits a single gaussian to an emission line
    at a user-specified wavelength. The input parameters are:
    
    s - the spectrum
    date - date the spectrum was taken (used for plotting and saving)
    line - the name of the emission line (used for plotting and saving)
    wave - the wavelength of the emission line
    cont - wavelength regions around the emission line used to fit the continuum [x1, x2, x3, x4]
    plot - true or false, tells the script whether or not to plot the results
    
    """
    
    #Cut region to the wavelength range of the region of interest
    #which is the bound by the continuum regions
    s = s[s[:,0]>cont[0]]
    s = s[s[:,0]<cont[3]]

    #Normalise flux - this helps the fitting function work correctly
    norm = np.median(s[:,1])
    s[:,1] /= norm 


    #Introduce the polynomial model for the continuum
    poly1 = PolynomialModel(degree=2, prefix='p1_')
    pars = poly1.guess(s[:,1], x=s[:,0])
    
    #Add the Gaussian model with some initial guesses
    gauss1 = GaussianModel(prefix='g1_')
    pars.update(gauss1.make_params())
    pars['g1_center'].set(wave, min=wave-4, max=wave+4)
    pars['g1_sigma'].set(0.5, min=0.1, max=2)

    #This is the model to be fit
    mod = poly1 + gauss1

    #Initialise the model
    init = mod.eval(pars, x=s[:,0])

    #Fit the model
    fit = mod.fit(s[:,1], pars, x=s[:,0])

    #Print and save results
    print('-------------------------------')
    print('Parameter    Value       Stderr')
    output = []
    for name, param in fit.params.items():
        out = f'{name:7s} {param.value:11.5f} {param.stderr:11.5f}'
        output.append(out)
        print (out)
    output.append('flux_norm	' + str(norm) + '	nan')
    
    #Un-comment to save fitting results
    #np.savetxt('../fits/'+date+'_'+line+'.txt', output, fmt='%s')

    if plot == True:
        #Plot
        plt.figure(figsize=(10,5))
        plt.title(date + ' ' + line)
    
        plt.plot(s[:,0], s[:,1], color='black')

        plt.plot(s[:,0], fit.best_fit, color='red', ls='--', lw=2)
        comps = fit.eval_components(x=s[:,0])
        plt.plot(s[:,0], comps['p1_'], '--', color='green', lw=0.8)
        plt.plot(s[:,0], comps['g1_'], '--', color='orange', lw=0.8)

        plt.xlabel(r'Rest Wavelength ($\AA$)')
        plt.ylabel('Normalised Flux')

        #Un-comment to save plot
        #plt.savefig(date + '_' + line + '.png')
        plt.show()

    #Save results
    g1_amp = fit.best_values['g1_amplitude']
    g1_sig = fit.best_values['g1_sigma']

    s[:,1] *= norm 

    return
```

Now we will try this function with an example spectrum, taken with the X-Shooter instrument on the Very Large Telescope (VLT). We will fit the OIII line at 5007A.

```python
spectrum = 'Data/TDE_spec_2.txt'

date = '20201215' #the spectrum was taken on this date

#Load the spectrum into a numpy array
s = np.genfromtxt(spectrum, comments='#')

#Correct for redshift
z = 0.01513
s[:,0] /= (1+z)

#Fit the OIII[5007] emission line
cont = [4990,5002,5013,5032]
OIII = fit_gauss(s, date, 'OIII_5007', 5007, cont, True)
```

The following function will fit the broad and narrow components of H$\alpha$ and the two Nitrogen lines simultaneously. Note that two gaussians are fitted to the broad H$\alpha$ component as it looks to be a double-peaked line.

```python
def halpha(s, date, cont, plot):

    line = 'h-alpha'

    #Cut region to the wavelength range of the region of interest
    #which is the bound by the continuum regions
    s = s[s[:,0]>cont[0]]
    s = s[s[:,0]<cont[3]]

    #Normalise flux
    norm = np.median(s[:,1])
    s[:,1] /= norm 

    #Set up polynomial, 2 broad Gaussians and 3 narrow Gaussians
    #The continuum
    poly1 = PolynomialModel(degree=2, prefix='p1_')
    pars = poly1.guess(s[:,1], x=s[:,0])
    
    #H-alpha broad 1
    bgauss1 = GaussianModel(prefix='bg1_')
    pars.update(bgauss1.make_params())
    pars['bg1_center'].set(6555, min=6552, max=6558)
    pars['bg1_sigma'].set(3, min=1, max=8)

    #H-alpha broad 2
    bgauss2 = GaussianModel(prefix='bg2_')
    pars.update(bgauss2.make_params())
    pars['bg2_center'].set(6572, min=6569, max=6573)
    pars['bg2_sigma'].set(3, min=1, max=8)

    #H-alpha narrow
    ngauss1 = GaussianModel(prefix='ng1_')
    pars.update(ngauss1.make_params())
    pars['ng1_center'].set(6563, min=6561, max=6565)
    pars['ng1_sigma'].set(1, min=0.1, max=2)
    #pars['ng1_amplitude'].set(1,min=0,max=10)

    #[NII]6584
    ngauss2 = GaussianModel(prefix='ng2_')
    pars.update(ngauss2.make_params())
    pars['ng2_center'].set(6584, min=6582, max=6585)
    pars['ng2_sigma'].set(1, min=0.1, max=2)
    #pars['ng2_amplitude'].set(1,min=0,max=10)

    #[NII]6549
    ngauss3 = GaussianModel(prefix='ng3_')
    pars.update(ngauss3.make_params())
    pars['ng3_center'].set(6549, min=6547, max=6551)
    pars['ng3_sigma'].set(expr='ng2_sigma') #fix width to other NII line
    pars['ng3_amplitude'].set(expr='ng2_amplitude/3') #3:1 ratio for NII lines

    #Fit data
    mod = poly1 + bgauss1 + bgauss2 + ngauss1 + ngauss2 + ngauss3

    init = mod.eval(pars, x=s[:,0])

    fit = mod.fit(s[:,1], pars, x=s[:,0])


    #Print and save results
    print('-------------------------------')
    print('Parameter    Value       Stderr')
    output = []
    for name, param in fit.params.items():
        out = f'{name:7s} {param.value:11.5f} {param.stderr:11.5f}'
        output.append(out)
        print (out)
    output.append('flux_norm	' + str(norm) + '	nan')
    
    #Un-comment to save
    #np.savetxt('.date+'_h-alpha_reg.txt', output, fmt='%s')

    if plot == True:
        #Plot
        plt.figure(figsize=(10,5))
        
        plt.title(date + ' H-alpha region')

        plt.plot(s[:,0], s[:,1], color='black')

        plt.plot(s[:,0], fit.best_fit, color='red', ls='-', lw=2)
        comps = fit.eval_components(x=s[:,0])
        plt.plot(s[:,0], comps['p1_'], '--', color='green', lw=0.8)
        
        plt.plot(s[:,0], comps['bg1_'], '--', color='orange', lw=0.8)
        plt.plot(s[:,0], comps['bg2_'], '--', color='orange', lw=0.8)

        plt.plot(s[:,0], comps['ng1_'], '--', color='magenta', lw=0.8)
        plt.plot(s[:,0], comps['ng2_'], '--', color='magenta', lw=0.8)
        plt.plot(s[:,0], comps['ng3_'], '--', color='magenta', lw=0.8)

        plt.xlabel('Rest Wavelength (A)')
        plt.ylabel('Normalised Flux')
        
        #Un-comment to save
        #plt.savefig('../fits/' + date + '_h-alpha_reg.png')
        plt.show()

    s[:,1] *= norm

    return
```

Now we input our spectrum into this function and see how it performs.

```python
#Fit our spectrum
cont_ha = [6507,6530,6599,6646]
ha = halpha(s, date, cont_ha, True)
```

Not a bad fit at all! The orange dashed lines are the broad components of H$\alpha$, the central magenta dashed line is narrow H$\alpha$ and the two lines either side are the NII lines.
