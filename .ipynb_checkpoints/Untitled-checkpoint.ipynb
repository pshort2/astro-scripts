In this notebook I will overplot different emission lines from the same spectrum in velocity space rather than wavelength space. This allows us to visually compare the velocity of the gas that is producing different emission lines, and determine whether or not the emission lines are being produced in the same region.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

First we import the spectrum as an array. Column 0 will be wavelength in units of Angstrom, and column 1 will be flux in ergs s$^{-1}$ cm$^{-2}$.

```python
spectrum = 'Data/TDE_spec_3.txt'

s = np.genfromtxt(spectrum,comments='#')
```

Now we will split the spectrum around the emission line regions we are interested in. This particular spectrum is from a tidal disruption event which is showing some curious emission lines from highly ionised Iron. Here we will pick out the lines [FeX] at 6375A and [FeXI] at 7892A. Note that the first step is to correct for redshift due to the expanding universe. Thankfully for the object in question the redshift is a known value.

```python
z = 0.01513

s[:,0] /= (1+z)

#[FeX]
sfx = s[s[:,0]>6200]
sfx = sfx[sfx[:,0]<6500]

#[FeXI]
sfxi = s[s[:,0]>7700]
sfxi = sfxi[sfxi[:,0]<8100]
```

Now we've isolated the regions of interest we can convert the wavelength to a velocity centered at the wavelength of the emission line. Because of the Doppler effect, light emitted by gas moving towards us gets blue-shifted to shorter wavelengths and light emitted by gas moving away from us gets red-shifted to longer wavelengths. The faster the gas is moving, the more the light gets shifted to longer/shorter wavelengths, and the broader the emission line gets.

We convert wavelength to velocity with the following:
\begin{equation}
v_i = \frac{\lambda_i - \lambda_{center}}{\lambda_{center}} \times c
\end{equation}

$v_i$: the velocity of each element in the array \
$\lambda_i$: the wavelength of each element in the array \
$\lambda_{center}$: the wavelength of the emission line \
$c$: the speed of light ($3\times10^5$km s$^{-1}$)

```python
sfx[:,0] = ((sfx[:,0]-6375)/6375)*3e5

sfxi[:,0] = ((sfxi[:,0]-7892)/7892)*3e5
```

Now we will plot the emission lines over the top of each other and see if their profiles are similar.

```python
plt.plot(sfx[:,0],sfx[:,1]/np.median(sfx[:,1]), label=r'[Fe X] $\lambda$6375',lw=0.8, alpha=0.7)
plt.plot(sfxi[:,0],sfxi[:,1]/np.median(sfxi[:,1]), label=r'[Fe XI] $\lambda$7892',lw=0.8, alpha=0.7)

plt.axvline(0, ls='--', color='black', alpha=0.5, lw=0.8)
plt.axvline(-302, ls='--', color='black', alpha=0.5, lw=0.8)
plt.axvline(260, ls='--', color='black', alpha=0.5, lw=0.8)

plt.xlim(-1000,1000)
plt.ylim(0.8, 1.35)
plt.legend()
plt.xlabel(r'Velocity (km s$^{-1}$)')
plt.ylabel('Normalised Flux')

plt.show()
```

What is interesting here is that the two emission lines have very similar profiles. I've used the black dashed lines to highlight the centre of the lines at velocity 0, as well as two peaks either side of the lines. The fact that these features appear for both the [FeX] and [FeXI] regions suggesting that they are real, and that we are seeing several regions of emitting gas, some moving towards as and some moving away.
